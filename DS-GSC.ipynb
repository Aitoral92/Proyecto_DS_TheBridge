```python
# AHora hacemos un join para volver a unir los dos dataframes por page, y tener la suma total de clicks e impresiones y la media de ctr y pos

esp=pd.merge(df_esp_sum, df_esp_mean, on='page', how='outer')
esp.sort_values(by='clicks', ascending=False)
```

```python
# Convertir las strings en objetos de fecha
first_day = datetime.strptime(start_day, "%Y-%m-%d")
last_day = datetime.strptime(end_day, "%Y-%m-%d")

# Calcular la diferencia entre las fechas
day_diff = last_day - first_day
```

```python
esp['mean clicks day']=esp['clicks']/(day_diff.days)
esp.sort_values(by='clicks', ascending=True).head()
```
